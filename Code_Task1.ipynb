```python
df4 = pd.read_excel(r"C:/Users/amaan/Documents/New folder/q4.xlsx")
df4.head()
```

```python
df4=df4.rename(columns={"Shares_By":"Name"})
```

```python
df4.head()
```

```python
#spliting column to get only numeric values of share_count
df4[['Share_Count','Share']] = df4.Total_Share.str.split(expand=True)
```

```python
df4.head()
```

```python
del df4["Total_Share"]
```

```python
df4.head()
```

```python
del df4["Share"]
```

```python
df4.head()
```

```python
df4["Share_Count"]=pd.to_numeric(df4["Share_Count"])
```

```python
df4["Share_Count"].dtype
```

```python
sorted_app3=df4.groupby('Name').agg({'Share_Count':'sum'}).sort_values(by='Share_Count',ascending=False)
```

```python
sorted_app4=df3.groupby('Name').agg({'Reactions_What':'sum'}).sort_values(by='Reactions_What',ascending=False)
```

```python
sorted_app4
```

```python
sorted_app5=df2.groupby('Name').agg({'Comment':'count'}).sort_values(by='Comment',ascending=False)
```

```python
sorted_app5
```

```python
result = pd.concat([sorted_app3, sorted_app4], axis=1, join="outer")
```

```python
result2=pd.concat([ result,sorted_app5], axis=1, join="outer")
```

```python
result2
```

```python
#filling NaN with 0 for calculation
result2=result2.fillna(0)
result2
```

```python
#counting total interaction
sum_interaction=result2["Share_Count"]+result2["Reactions_What"]+result2["Comment"]
```

```python
result2["sum_interaction"]=sum_interaction
```
