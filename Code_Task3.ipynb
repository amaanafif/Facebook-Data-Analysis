```python
##loading required libraries
import pandas as pd
import numpy as np
import os
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
#importing excel file
df1 = pd.read_excel(r"C:/Users/amaan/Documents/New folder/q2.xlsx")
df1.head()
```

```python
#Finding NaN values
df1.isna().sum()
```

```python
#Droping NaN values
df1.dropna(subset = ["Comment"], inplace=True)
```

```python
df1.isna().sum()
```

```python
#converting everything to string
df1["Comment"] = df1["Comment"].astype(str)
```

```python
df1["Comment"].dtype
```

```python
text = " ".join(df1.Comment)
```

```python
#removing special characters
import re
new_txt=''
for k in text.split("\n"):
    new_txt+=re.sub(r"[^a-zA-Z0-9]+", ' ', k)
```

```python
#creating wordcloud
word_cloud = WordCloud(collocations = False, background_color = 'white').generate(new_txt)
```

```python
#plotting wordcloud
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
